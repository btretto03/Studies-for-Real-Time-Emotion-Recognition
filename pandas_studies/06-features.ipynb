```python
import pandas as pd
import numpy as np
```

```python
my_df = pd.read_csv('ShelterDogs5.csv')
my_df
```

```python
# Create a new column
my_df["Frame_Header"] = ["Dog 1", "Dog 2", "Dog 3", "Dog 4", "Dog 5"]
my_df
```

```python
# Set the Index
my_df.set_index("Frame_Header", inplace=True)
# Note: Using inplace=True makes the change permanent
my_df
```

```python
# Reset the Index
my_df.reset_index(inplace=True)
my_df
```

```python
# Drop the extra column
my_df.drop("Frame_Header", axis=1, inplace=True)
my_df
```
