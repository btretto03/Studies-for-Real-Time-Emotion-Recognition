```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
my_df = pd.DataFrame(randn(100, 4), columns=["Monday", "Tuesday", "Wednesday", "Thursday"])
my_df
```

```python
# Histogram
my_df['Wednesday'].hist()
```

```python
# Histogram with smaller bins
my_df['Wednesday'].hist(bins = 100)
```

```python
# Another options
my_df['Wednesday'].plot(kind = "hist")
```

```python
my_df['Wednesday'].plot(kind = "hist", grid = False)
```

```python
my_df['Wednesday'].plot(kind = "hist", legend = True)
```

```python
# Area Plot
my_df.abs().plot(kind='area')
```

```python
# Option 1
my_df.plot(kind='area', stacked=False)
```

```python
# Option 2 Transparency
my_df.abs().plot(kind='area', alpha=0.3)
```

```python
# Add title
my_df.abs().plot(kind='area', alpha=0.3, title="My Awesome Area Plot")
```

```python
# Remove legend
my_df.abs().plot(kind='area', alpha=0.3, legend=False)
```

```python
# Just one specific column 
my_df['Monday'].abs().plot(kind='area')
```

```python
# Plot multiple columns
my_df[['Monday', 'Thursday']].abs().plot(kind='area')
```

```python
# Add grid
my_df.abs().plot(kind='area', grid=True)
```
