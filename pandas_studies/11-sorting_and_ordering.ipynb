```python
import pandas as pd
import numpy as np
```

```python
# Create Data
stuff = {
    'Corporation': ['Apple', 'Google', 'Meta', 'Apple', 'Google', 'Meta'],
    'Employees': ['John', 'April', 'Wes', 'Beth', 'Justin', 'Steph'],
    'Salary': [200, 220, 190, 130, 120, 150]
}
my_df = pd.DataFrame(stuff)
my_df
```

```python
# Sort by Salary (Ascending)
my_df.sort_values('Salary')
```

```python
# Sort by Salary (Descending)
my_df.sort_values('Salary', ascending=False)
```

```python
# Permanent Change
my_df.sort_values('Salary', ascending=False, inplace=True)
my_df
```

```python
# Sort strings alphabetically
my_df.sort_values('Corporation')
```

```python
# Sort strings descending
my_df.sort_values('Corporation', ascending=False)
```
