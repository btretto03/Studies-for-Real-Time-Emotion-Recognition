```python
import torch
```

```python
my_torch = torch.arange(10)
my_torch
```

```python
# Reshape and view
my_torch = my_torch.reshape(2, 5)
my_torch
```

```python
# Reshape if we don't know the number of items using -1
my_torch2 = torch.arange(10)
my_torch2 = my_torch2.reshape(5,-1)
my_torch2
```

```python
# With reshape and view they will update
my_torch3 = torch.arange(10)
my_torch3
```

```python
my_torch4 = my_torch3.reshape(2,5)
my_torch4
```

```python
my_torch3[0] = 1982
my_torch3
```

```python
my_torch4 # it changes too
```

```python
# Slices
my_torch5 = torch.arange(10)
my_torch5[7] # note it's not just 7 is tensor(7)
```

```python
# Grab slices
my_torch6 = my_torch5.reshape(5, 2)
my_torch6[:, 1]
```
