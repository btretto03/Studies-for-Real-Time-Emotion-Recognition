```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1, 2, 3, 4])
tensor_b = torch.tensor([5, 6, 7, 8])
```

```python
# Shorthand addition
tensor_a + tensor_b
```

```python
# Longhand addition
add_long = torch.add(tensor_a, tensor_b)
add_long
```

```python
# Shorthand subtraction
tensor_b - tensor_a
```

```python
# Longhand subtraction
sub_long = torch.sub(tensor_b, tensor_a)
sub_long
```

```python
# Shorthand multiplication
tensor_a * tensor_b
```

```python
# Longhand multiplication
mul_long = torch.mul(tensor_a, tensor_b)
mul_long
```

```python
# Shorthand division
tensor_b / tensor_a
```

```python
# Longhand division
div_long = torch.div(tensor_b, tensor_a)
div_long
```

```python
# Shorthand modulus
tensor_b % tensor_a
```

```python
# Longhand remainder
mod_long = torch.remainder(tensor_b, tensor_a)
mod_long
```

```python
# Longhand for powers (tensor_a to the power of tensor_b)
pow_long = torch.pow(tensor_a, tensor_b)
pow_long
```

```python
# Note that this creates a NEW tensor in memory and leaves tensor_a unchanged
tensor_a
```

```python
# Using the underscore (_) at the end of the function
# This modifies tensor_a DIRECTLY and returns nothing or the modified tensor
tensor_a.add_(tensor_b)
tensor_a
```
