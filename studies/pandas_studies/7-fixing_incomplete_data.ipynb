```python
import pandas as pd
import numpy as np
```

```python
# Create data with a missing value
stuff = {'A':[1,2,3], 'B':[4,np.nan,6], 'C':[7,8,9], 'D':[10,11,12]}
my_df = pd.DataFrame(stuff)
my_df
```

```python
# Drop rows with any null values (Default is axis=0)
my_df.dropna()
```

```python
# Drop columns with any null values
my_df.dropna(axis=1)
```

```python
# Not permanent, implace = false
my_df
```

```python
# with more than 1 null value
stuff = {'A':[1,2,3], 'B':[4,np.nan,np.nan], 'C':[7,8,9], 'D':[10,11,12]}
my_df2 = pd.DataFrame(stuff)
my_df2
```

```python
# Set a threshold (Keep rows/columns with at least 1 non-null values)
my_df2.dropna(thresh=1, axis=1)
```

```python
# Set a threshold (Keep rows/columns with at least 2 non-null values)
my_df2.dropna(thresh=2, axis=1)
```

```python
# Fill null values with a specific word or number
my_df2.fillna(value=5, inplace=True)
```
