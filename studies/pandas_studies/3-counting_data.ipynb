```python
import pandas as pd
import numpy as np
```

```python
my_df = pd.read_csv('ShelterDogs.csv')
```

```python
# Count distinct values - descending
my_df['name'].value_counts()
```

```python
# Count distinct values - ascending
my_df['name'].value_counts(ascending = True)
```

```python
my_df['housebroken'].value_counts() # but we have same with NaN(none)
```

```python
my_df['housebroken'].value_counts(dropna = False) # now we see all of them
```

```python
# Get relative frequency (percentage)
my_df['name'].value_counts(normalize = True) 
```

```python
# Get specific item
print(my_df['name'].value_counts()['Lili'])
```
