```python
import pandas as pd
import numpy as np
```

```python
my_df = pd.read_csv('ShelterDogs5.csv')
```

```python
# Add column from list
Test = ["A", "A", "B", "D", "D"]
my_df["Test"] = Test
my_df
```

```python
# Add default values
my_df["Alive/Dead"] = [True] * len(my_df)
my_df
```

```python
# Add NaN values
my_df["Happy?"] = [np.nan] * len(my_df)
my_df
```

```python
# Add column with .insert()
my_df.insert(1, "Adopted", ["YES"]* len(my_df), True)
my_df
```

```python
# Add column with .assign(), it creates a new df
my_df2 = my_df.assign(HORSE = [False]*len(my_df))
my_df2
```

```python
# Remove column(axis = 1)
my_df2.drop("HORSE", axis=1)
```

```python
my_df2 
```

```python
my_df2.drop("HORSE", axis=1, inplace = True) # to remove permanently we need the bool inplace
my_df2
```

```python
# Removing Row
my_df2.drop(3") # the pattern is axis = 0(row)
```
