```python
import pandas as pd
import numpy as np
```

```python
my_df = pd.read_csv('ShelterDogs5.csv')
```

```python
# Grab row method 1: Location
my_df.loc[2] # If it was a string just put in ''
```

```python
# Grab row method 2: Index location
my_df.iloc[3]
```

```python
# Specific points(row and column) 
my_df.loc[3, 'color']
```

```python
# Subsets
# my_df.loc[['row', 'row'], ['column', 'column']]
my_df.loc[[1, 3], ["color", "size"]]
```

```python
# Conditional selection, it returns booleans
my_df == 'yellow-brown'
```

```python
# Returns Dataframes with data
my_df[my_df == 'yellow-brown']
```

```python
# Run them on a column
my_df[my_df == 'yellow-brown']['color']
```

```python
# Return multiple columns
my_df[my_df == 'yellow-brown'][['color','name']]
# As here we only have strings, we cannot use comparatives(>, <, ...) if it was numeros we could use it
```

```python
# Multiple condition AND(&)
my_df[(my_df['color'] == 'yellow-brown') & (my_df['breed'] == 'Unknown Mix')]
```

```python
# Get lenght
len(my_df[(my_df['color'] == 'yellow-brown') & (my_df['breed'] == 'Unknown Mix')])
```

```python
# Multiple condition or(|)
my_df[(my_df['color'] == 'yellow-brown') | (my_df['breed'] == 'Unknown Mix')]
```

```python
# Get lenght
len(my_df[(my_df['color'] == 'yellow-brown') | (my_df['breed'] == 'Unknown Mix')])
```

```python
# Return specific item
my_df[(my_df['color'] == 'yellow-brown') | (my_df['breed'] == 'Unknown Mix')]['name']
```
