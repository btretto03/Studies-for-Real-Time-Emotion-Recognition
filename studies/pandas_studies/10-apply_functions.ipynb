```python
import pandas as pd
import numpy as np
```

```python
# Create Data
stuff = {
    'Corporation': ['Apple', 'Google', 'Meta', 'Apple', 'Google', 'Meta'],
    'Employees': ['John', 'April', 'Wes', 'Beth', 'Justin', 'Steph'],
    'Salary': [200, 220, 190, 130, 120, 150]
}
```

```python
my_df = pd.DataFrame(stuff)
my_df
```

```python
# Creates function
def times1000(x):
    return x * 1000
```

```python
# Use apply function
my_df['Salary'].apply(times1000)
```

```python
# Make it into a Dataframe
pd.DataFrame(my_df['Salary'].apply(times1000))
```

```python
# Namer Function
def namer(x):
    if x == 'John':
        return 'John Elder'
    else:
        return x
```

```python
# Apply to Employees column
my_df['Employees'].apply(namer)
```

```python
# Use Lambda instead
my_df['Salary'].apply(lambda x: x * 1000)
```

```python
# Lambda with formatting
my_df['Salary'].apply(lambda x: f"{x * 1000:,d}")
```
