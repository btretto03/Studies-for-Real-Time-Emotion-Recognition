```python
import pandas as pd
import numpy as np
```

```python
# Pandas series - 1D array with any type of data
```

```python
my_series = [1, 5, 9, 7]
my_var = pd.Series(my_series)
my_var
```

```python
# Print specific value
```

```python
print(my_var[2]) # When you use print it forces Jupyter to print just the value
my_var[2]
```

```python
# Labels using an index argument
```

```python
my_index = ["A", "B", "C", "D"]
my_var2 = pd.Series(my_series, my_index)
my_var2
```

```python
print(my_var2["B"])
```

```python
# Key value dictionary
```

```python
cars = {"Tesla": 12, "Mercedes": 42, "BMW": 3}
my_var3 = pd.Series(cars)
my_var3
```
